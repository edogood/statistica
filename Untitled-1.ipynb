```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
numeri_generati = []
n=10000000
media, dev_std = 25, 1.5
for i in range(n):
    nuovo_numero = np.random.normal(media, dev_std)
    numeri_generati.append(nuovo_numero)
    
plt.hist(numeri_generati, bins=30, edgecolor='black')
plt.show()
```

```python

# Parametri
n_stud = 10000
n_esami = 20
min_media_voto, max_media_voto = 15, 30
min_std_dev, max_std_dev = 0.1, 8
n_tentativi_max = 150

# Creazione studenti con media e deviazione standard casuali
stud = np.array([[np.random.uniform(min_media_voto, max_media_voto), 
                  np.random.uniform(min_std_dev, max_std_dev)] 
                 for _ in range(n_stud)])

# Inizializzazione delle liste
esito_esami = [[] for _ in range(n_stud)]
tentativi = np.zeros(n_stud, dtype=int)

# Simulazione degli esami
for s, (media, std_dev) in enumerate(stud):
    while len(esito_esami[s]) < n_esami:
        tentativi[s] += 1
        voto = np.random.normal(media, std_dev)
        voto = min(30, round(voto))  # Arrotonda e limita a 30
        if voto >= 18:
            esito_esami[s].append(voto)
        if tentativi[s] > n_tentativi_max:
            break

# Grafico della distribuzione dei voti
tutti_voti = [voto for esiti in esito_esami for voto in esiti]
plt.hist(tutti_voti, bins=12, edgecolor='black')
plt.xlabel('Voto')
plt.ylabel('Frequenza')
plt.title('Distribuzione dei voti')
plt.show()

# Calcolo della media per ogni studente
media_esiti = np.array([np.mean(esito) if esito else 0 for esito in esito_esami])

# Normalizzazione della deviazione standard per i colori
std_dev_list = (stud[:, 1] - min_std_dev) / (max_std_dev - min_std_dev)

# Scatter plot Tentativi vs Media Voti (con y  18)
plt.scatter(tentativi, media_esiti, c=std_dev_list, cmap="viridis")
plt.xlabel('Tentativi')
plt.ylabel('Media voti')
plt.ylim(18, 30)  # Imposta il range della y tra 18 e 30
plt.colorbar(label='Deviazione Standard Normalizzata')
plt.title('Relazione tra tentativi e media voti')
plt.show()


# Parametri
n_stud = 10000
n_esami = 20
min_media_voto, max_media_voto = 24.99, 25
min_std_dev, max_std_dev = 0.1, 8
n_tentativi_max = 150

# Creazione studenti con media e deviazione standard casuali
stud = np.array([[np.random.uniform(min_media_voto, max_media_voto), 
                  np.random.uniform(min_std_dev, max_std_dev)] 
                 for _ in range(n_stud)])

# Inizializzazione delle liste
esito_esami = [[] for _ in range(n_stud)]
tentativi = np.zeros(n_stud, dtype=int)

# Simulazione degli esami
for s, (media, std_dev) in enumerate(stud):
    while len(esito_esami[s]) < n_esami:
        tentativi[s] += 1
        voto = np.random.normal(media, std_dev)
        voto = min(30, round(voto))  # Arrotonda e limita a 30
        if voto >= 18:
            esito_esami[s].append(voto)
        if tentativi[s] > n_tentativi_max:
            break

# Grafico della distribuzione dei voti
tutti_voti = [voto for esiti in esito_esami for voto in esiti]
plt.hist(tutti_voti, bins=12, edgecolor='black')
plt.xlabel('Voto')
plt.ylabel('Frequenza')
plt.title('Distribuzione dei voti')
plt.show()

# Calcolo della media per ogni studente
media_esiti = np.array([np.mean(esito) if esito else 0 for esito in esito_esami])

# Normalizzazione della deviazione standard per i colori
std_dev_list = (stud[:, 1] - min_std_dev) / (max_std_dev - min_std_dev)

# Scatter plot Tentativi vs Media Voti (con y  18)
plt.scatter(tentativi, media_esiti, c=std_dev_list, cmap="viridis")
plt.xlabel('Tentativi')
plt.ylabel('Media voti')
plt.ylim(18, 30)  # Imposta il range della y tra 18 e 30
plt.colorbar(label='Deviazione Standard Normalizzata')
plt.title('Relazione tra tentativi e media voti')
plt.axvline(x=40, c= "k", linestyle="--", label="Due tent.per esame")
plt.axvline(x=60, c= "r", linestyle="--", label="Tre tent.per esame")
plt.axvline(x=80, c= "g", linestyle="--", label="Quattro tent.per esame")
plt.legend()
plt.show()
plt.show()


# Grafico a barre delle medie (con y  18)
plt.bar(range(n_stud), media_esiti, color='blue', alpha=0.1)
plt.xlabel('Studente')
plt.ylabel('Media Voti')
plt.ylim(18, 30)  # Imposta il range della y tra 18 e 30
plt.title('Media dei voti per studente')


plt.show()
# Grafico a barre delle medie (con y  18)
plt.bar(range(n_stud), media_esiti, color='blue', alpha=0.1)
plt.xlabel('Studente')
plt.ylabel('Media Voti')
plt.ylim(18, 30)  # Imposta il range della y tra 18 e 30
plt.title('Media dei voti per studente')


plt.show()
```
